```python
import pandas as pd 
```

```python
#Importing Diversity Index Dataset 
diversity_index = "Resources/us_diversity_index.csv"
diversity_index_df = pd.read_csv(diversity_index)
diversity_index_df.head()
```

```python
#Spliting location in to County and State in the diversity dataset 
split_diversity_df = diversity_index_df
split_diversity_df['County'], split_diversity_df['State'] = diversity_index_df['Location'].str.rsplit(', ').str
#Showing results of the split 
split_diversity_df.head()
```

```python
#Grouping each race by "US States" & "US Counties" and calculating the mean 

group_diversity_df = pd.DataFrame({"Diversity-Index":split_diversity_df.groupby(["State","County"])["Diversity-Index"].mean(),
                       "Black":split_diversity_df.groupby(["State","County"])["Black or African American alone, percent, 2013"].mean(),
                       "American Indian":split_diversity_df.groupby(["State","County"])["American Indian and Alaska Native alone, percent, 2013"].mean(),
                       "Asian":split_diversity_df.groupby(["State","County"])["Asian alone, percent, 2013"].mean(),
                       "Native Hawaiian":split_diversity_df.groupby(["State","County"])["Native Hawaiian and Other Pacific Islander alone, percent,"].mean(),
                       "Two or More Races":split_diversity_df.groupby(["State","County"])["Two or More Races, percent, 2013"].mean(),
                       "Hispanic/ Latino":split_diversity_df.groupby(["State","County"])["Hispanic or Latino, percent, 2013"].mean(),
                       "White":split_diversity_df.groupby(["State","County"])["White alone, not Hispanic or Latino, percent, 2013"].mean()})
```

```python
#Showing the new dataframe that groups columns by "states" and "counties"
group_diversity_df.head(20)
```

```python
#Importing Unemployment Rate Dataset
us_unemployment_df = pd.read_csv('Resources/us_enemployment_state_county.csv')
us_unemployment_df.head()
```

```python
#Dropping "Year" and "Month" columns as we don't need it and only selecting "State", "County" & "Rate"
us_unemployment_df = us_unemployment_df[["County", "State", "Rate"]].copy()
us_unemployment_df.head()
```

```python
#Creating abbreviations for "States" as the states are written in full 
abbreviate_us_states = {
   'Alabama': 'AL',
   'Alaska': 'AK',
   'Arizona': 'AZ',
   'Arkansas': 'AR',
   'California': 'CA',
   'Colorado': 'CO',
   'Connecticut': 'CT',
   'Delaware': 'DE',
   'Florida': 'FL',
   'Georgia': 'GA',
   'Hawaii': 'HI',
   'Idaho': 'ID',
   'Illinois': 'IL',
   'Indiana': 'IN',
   'Iowa': 'IA',
   'Kansas': 'KS',
   'Kentucky': 'KY',
   'Louisiana': 'LA',
   'Maine': 'ME',
   'Maryland': 'MD',
   'Massachusetts': 'MA',
   'Michigan': 'MI',
   'Minnesota': 'MN',
   'Mississippi': 'MS',
   'Missouri': 'MO',
   'Montana': 'MT',
   'Nebraska': 'NE',
   'Nevada': 'NV',
   'New Hampshire': 'NH',
   'New Jersey': 'NJ',
   'New Mexico': 'NM',
   'New York': 'NY',
   'North Carolina': 'NC',
   'North Dakota': 'ND',
   'Ohio': 'OH',
   'Oklahoma': 'OK',
   'Oregon': 'OR',
   'Pennsylvania': 'PA',
   'Rhode Island': 'RI',
   'South Carolina': 'SC',
   'South Dakota': 'SD',
   'Tennessee': 'TN',
   'Texas': 'TX',
   'Utah': 'UT',
   'Vermont': 'VT',
   'Virginia': 'VA',
   'Washington': 'WA',
   'West Virginia': 'WV',
   'Wisconsin': 'WI',
   'Wyoming': 'WY',
}
```

```python
#Changing full states name to abbreviated state names in the dataframe 
abbrev_unemploy_df = us_unemployment_df
for key, value in abbreviate_us_states.items():
    abbrev_unemploy_df.loc[abbrev_unemploy_df['State']==key, 'State'] = value
abbrev_unemploy_df.head()
```

```python
#Grouping the dataframe by state and changing "Rate" column name to "Unemployment Rate" and calculating the mean 
group_unemploy_df = pd.DataFrame({"Unemployment Rate":abbrev_unemploy_df.groupby(["State","County"])["Rate"].mean()})
group_unemploy_df.head(20)
```

```python
#Importing Median Income Dataset
us_median_income = pd.read_csv('Resources/us_states_counties_median_income.csv')
us_median_income.head()
```

```python
#Selecting usable columns only and renaming "State Code" to "State"
us_median_income = us_median_income[['County',"State Code", 'Population',"Median household income"]]
us_median_income = us_median_income.rename(columns={"State Code": "State"})
us_median_income.head()
```

```python
#grouping the median income by state and calculating the mean 
group_us_median_income = pd.DataFrame({"Median household income":us_median_income.groupby(["State","County"])["Median household income"].mean(),
                              "Population":us_median_income.groupby(["State","County"])["Population"].sum()})
group_us_median_income.head()
```

```python
#Merging tables
#Merging unemployment rate with median income table 
unemploy_medianinc_merged = pd.merge(group_unemploy_df,group_us_median_income, on=["State",'County'],how="inner")
unemploy_medianinc_merged.head()
```

```python
#Merging unemployment rate, median income and diversity index tables 
unemploy_medianinc_diversity_merge = pd.merge(unemploy_medianinc_merged,group_diversity_df, on=["State",'County'],how="inner")
unemploy_medianinc_diversity_merge.head()
```

```python
#Restoring State and county back to columns from the group format 
merged_final = unemploy_medianinc_diversity_merge.reset_index()
merged_final.head(20)
```
